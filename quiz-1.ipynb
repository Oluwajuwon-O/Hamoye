```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv('energydata_complete.csv')
```

```python
data.isnull().sum()
```

```python
columns = ['Appliances','T1','RH_1','T2','RH_2','T3','RH_3','T4',
          'RH_4','T5','RH_5','T6','RH_6','T7','RH_7','T8','RH_8',
          'T9','RH_9','T_out','Pressure','RH_out','Windspeed',
           'Visibility','Tdewpoint','rv1','rv2']
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

df = data.drop(columns = ['date','lights'])
df = pd.DataFrame(scaler.fit_transform(df), columns = columns)
df
```

```python
features = df.drop(columns = ['Appliances'])
features
```

```python
target = df['Appliances']
target
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(features,target,test_size = 0.3,random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
# comparing effects of regularisation

def get_weights_df(model,feat,col_name):
    # this function returns the weight of every feature
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    
    return weights_df
```

```python
linear_model_weights = get_weights_df(lr, x_train, 'Linear_Model_Weight')
linear_model_weights
```

```python
from sklearn.linear_model import Ridge

ridge_reg = Ridge(alpha = 0.4)
ridge_reg.fit(x_train, y_train)
```

```python
# Root mean square error (RMSE)
# RMSE -> square-root of mean squared error

from sklearn.metrics import mean_squared_error
mse = (mean_squared_error(y_test, y_pred))
print ('mean squared error: ' f'{mse}')
rmse = np.sqrt(mse).round(decimals = 2)
print ('root mean square error: ' f'{rmse}')
```

```python
from sklearn.linear_model import Lasso

lasso_reg = Lasso(alpha = 0.001)
lasso_reg.fit(x_train,y_train)
```

```python
lasso_weights_df = get_weights_df(lasso_reg,  x_train, 'Lasso_Weight')
lasso_weights_df.round(decimals = 3)
```

```python
# Root mean square error (RMSE)
# RMSE -> square-root of mean squared error

from sklearn.metrics import mean_squared_error
mse = (mean_squared_error(y_test, y_pred)).round(decimals = 3)
print ('mean squared error: ' f'{mse}')
rmse = np.sqrt(mse).round(decimals = 3)
print ('root mean square error: ' f'{rmse}')
```
